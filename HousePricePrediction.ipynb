```python
# Defining a function to plot bathrooms for different bedroom configurations
def plot_bedrooms(df):
    for i in np.sort(df['bedrooms'].unique()):
        i_bed = df[df['bedrooms'] == i]
        sns.countplot(x = i_bed['bathrooms'])
        plt.title(f"{i} Bedroom/s")
        plt.show()

# Creating a copy of the modified Lagos DataFrame
lagos_df2 = lagos_df1.copy()

# Applying the function to plot bathrooms for different bedroom configurations
plot_bedrooms(lagos_df2)
```

```python
# Defining a function to remove bathroom outliers based on bedroom configuration limits
def remove_bathroom_outliers(df):
    bath_limits = {
        1: (1, 2),
        2: (1, 3),
        3: (1, 4),
        4: (2, 5),
        5: (3, 6),
        6: (4, 7),
        7: (5, 8),
        8: (5, 9),
        9: (5, 9)
    }
    for index, row in df.iterrows():
        bedrooms = row['bedrooms']
        bathrooms = row['bathrooms']
        min_bath, max_bath = bath_limits.get(bedrooms, (0, float('inf')))
        if not min_bath <= bathrooms <= max_bath:
            df.drop(index, inplace=True)
    return df

# Creating a copy of the modified Lagos DataFrame and removing bathroom outliers
lagos_df3 = lagos_df2.copy()
remove_bathroom_outliers(lagos_df3)
```

```python
# Applying the function to plot bathrooms for different bedroom configurations after removing outliers
plot_bedrooms(lagos_df3)
```

```python
# Defining a function to plot toilets for different bedroom configurations
def plot_toilets(df):
    for i in np.sort(df['toilets'].unique()):
        i_toilets = df[df['bedrooms'] == i]
        sns.countplot(x = i_toilets['toilets'])
        plt.title(f"{i} Toilets/s")
        plt.show()

# Applying the function to plot toilets for different bedroom configurations after removing outliers
plot_toilets(lagos_df3)
```

```python
# Defining a function to remove toilet outliers based on bedroom configuration limits
def remove_toilet_outliers(df):
    toilet_limits = {
        1: (1, 2),
        2: (1, 3),
        3: (1, 4),
        4: (2, 5),
        5: (3, 6),
        6: (4, 7),
        7: (5, 8),
        8: (5, 9),
        9: (5, 9)
    }
    for index, row in df.iterrows():
        bedrooms = row['bedrooms']
        toilets = row['toilets']
        min_toilet, max_toilet = toilet_limits.get(bedrooms, (0, float('inf')))
        if not min_toilet <= toilets <= max_toilet:
            df.drop(index, inplace=True)
    return df

# Creating a copy of the DataFrame and removing toilet outliers
lagos_df4 = lagos_df3.copy()
remove_toilet_outliers(lagos_df4)
```

```python
# Applying the function to plot toilets for different bedroom configurations after removing outliers
plot_toilets(lagos_df4)
```

```python
# Defining a function to remove parking space outliers based on bedroom configuration limits
def remove_parking_outliers(df):
    parking_limits = {
      1: (1, 2),
      2: (1, 3),
      3: (1, 4),
      4: (2, 5),
      5: (3, 6),
      6: (4, 7),
      7: (5, 8),
      8: (5, 9),
      9: (5, 9)
    }
    for index, row in df.iterrows():
        bedrooms = row['bedrooms']
        parking = row['parking_space']
        min_parking, max_parking = parking_limits.get(bedrooms, (0, float('inf')))
        if not min_parking <= parking <= max_parking:
            df.drop(index, inplace=True)
    return df

# Creating a copy of the DataFrame and removing parking space outliers
lagos_df5 = lagos_df4.copy()
remove_parking_outliers(lagos_df5)
```

```python
# Creating a copy of the DataFrame and filtering rows with prices greater than or equal to 3 million
lagos_df6 = lagos_df5.copy()
lagos_df6 = lagos_df6[lagos_df6['price'] >= 3000000]
```

```python
# Visualizing the correlation between features in the DataFrame using a heatmap
sns.heatmap(lagos_df6.corr(), annot=True)
plt.show()
```

```python
# Dropping specific columns from the DataFrame
lagos_df7 = lagos_df6.drop(['toilets', 'title', 'state'], axis=1)
```

```python
# Creating dummy variables for the 'town' column and concatenating them with the DataFrame
dummies = pd.get_dummies(lagos_df6['town'])
lagos_df8 = pd.concat([lagos_df7, dummies], axis=1)
lagos_df8 = lagos_df8.drop('town', axis=1)
```

```python
# Separating features (X) and the target variable (y) from the DataFrame
X = lagos_df8.drop('price', axis=1)
y = lagos_df8['price']
```

```python
# Converting X and y dataframes into numpy arrays
X = np.array(X)
y = np.array(y)
```

```python
# Create a Pipeline and Standardize X and y by scaling them 
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

model = make_pipeline(StandardScaler())
model.fit(X, y)
```

```python
# Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Importing machine learning models for regression & Cross Validation 
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
# Initializing and training the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Calculating R^2 score for Linear Regression model
lr_model_score = lr_model.score(X_test, y_test)
```

```python
# Initializing and training the XGBoost Regressor model
xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)

# Calculating R^2 score for XGBoost Regressor model
xgb_model_score = xgb_model.score(X_test, y_test)
```

```python
# Initializing and training the Random Forest Regressor model
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

# Calculating R^2 score for Random Forest Regressor model
rf_model_score = rf_model.score(X_test, y_test)
```

```python
# Performing cross-validation for Linear Regression model
lr_cv_scores = cross_val_score(lr_model, X, y, cv=5)
lr_cv_mean = np.mean(lr_cv_scores)

# Performing cross-validation for XGBoost Regressor model
xgb_cv_scores = cross_val_score(xgb_model, X, y, cv=5)
xgb_cv_mean = np.mean(xgb_cv_scores)

# Performing cross-validation for Random Forest Regressor model
rf_cv_scores = cross_val_score(rf_model, X, y, cv=5)
rf_cv_mean = np.mean(rf_cv_scores)
```

```python
# Creating a dataframe to display model evaluation metrics
eval_metrics = pd.DataFrame({
    'Model': ['Linear Regression', 'XGBoost Regressor', 'Random Forest Regressor'],
    'R^2 Score': [lr_model_score, xgb_model_score, rf_model_score],
    'Cross-Validation Mean': [lr_cv_mean, xgb_cv_mean, rf_cv_mean]
})
eval_metrics
```

```python
#Lets create a list of the independent columns to create xcolumn variable

xcolumns = lagos_df8.columns.drop('price')
xcolumns
```

```python
# Defining a function to predict house prices using the trained models
def predict_price(model, town, bedrooms, bathrooms, parking_space):
    town_index = np.where(xcolumns == town)[0][0]
    x_input = np.zeros(len(xcolumns))
    x_input[0] = bedrooms
    x_input[1] = bathrooms
    x_input[2] = parking_space
    if town_index >= 0:
        x_input[town_index] = 1
    return model.predict([x_input])[0]
```

```python
predict_price(rf_model, 'Lekki', 4, 5, 3)
```

```python
from sklearn.model_selection import learning_curve

# Plot learning curves for Random Forests model
train_sizes, train_scores, val_scores = learning_curve(RandomForestRegressor(random_state=42), X_train, y_train, cv=5, scoring='r2')

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Training Score')
plt.plot(train_sizes, np.mean(val_scores, axis=1), label='Validation Score')
plt.xlabel('Training Set Size')
plt.ylabel('R-squared Score')
plt.title('Learning Curves - Random Forests Model')
plt.legend()
plt.show()
```
